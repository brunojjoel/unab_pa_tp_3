```python
class Punto:

    def __init__(self, x, y):
        self.x = x
        self.y = y

    def eje_x(self):
        return self.y == 0

    def eje_y(self):
        return self.x == 0

    def impresion(self):
        return f"Punto({self.x}, {self.y})"

    def opuesto(self):
        return Punto(-self.x, -self.y)
```

```python
p = Punto(1, 8)

print(p.impresion())
print("En eje X:", p.eje_x())
print("En eje Y:", p.eje_y())

p_opuesto = p.opuesto()
print("Opuesto:", p_opuesto.impresion())
```
